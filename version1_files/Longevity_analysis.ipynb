```python
import matplotlib as plt
import pandas as pd
from plotnine import *
```

```python
# Reading in Data

albums_df = pd.read_csv("combined_artists.csv")

songs_df = pd.read_csv("updated_combined_artists.csv")
```

```python
songs_df
```

```python
# Calculating Longevity for Artist

# Convert 'release_date' column to datetime format
songs_df['release_date'] = pd.to_datetime(songs_df['release_date'])

# Group by 'artist' and find the first and last release dates
artist_dates = songs_df.groupby('artist')['release_date'].agg(['min', 'max']).reset_index()

# Calculate the time difference
artist_dates['time_difference'] = (artist_dates['max'] - artist_dates['min'])

# Extract the years from the timedelta and convert to numeric
artist_dates['time_difference_years'] = artist_dates['time_difference'].dt.days // 365

# Display the result
print(artist_dates[['artist', 'min', 'max', 'time_difference_years']])
```

```python
# Merge the two DataFrames on the 'artist' column
merged_df = pd.merge(songs_df, artist_dates[['artist', 'time_difference_years']], on='artist', how='left')

merged_df['time_difference_years'] = merged_df['time_difference_years'].rename('longevity') 
```
